```python
# imports
from googletrans import Translator, constants
from pprint import pprint
```

```python
# instantiate translator class
translator = Translator()
```

```python
# print all available languages
lang_dict = {v: k for k, v in constants.LANGUAGES.items()}
```

```python
# create a list from the dictionary to add to my app
langauge_list = []
for key, value in lang_dict.items():
  langauge_list.append(key.capitalize())

langauge_list
```

```python
# in order to transform the prediction output from a number, I need to create a dictionary that will line up with the actual words
pred_dict = {0: 'bed', 1: 'cat', 2: 'chair'}
```

```python
def prediction(pred):

  """My app will predict a word based on the image provided. This function will take the prediction and find the associate word in the dictionary created above."""

  word = pred_dict[pred]
  print(word)
```

```python
# test prediction funtion
prediction(0)
```

```python
def translate(word, language):

  """ This function takes two arguments (word & language) and feeds them through the Google translator."""

  language = language.lower()
  translation = translator.translate(word, dest=lang_dict[language])
  return(f"{(translation.origin).replace('_', ' ')} ({translation.src}) --> {translation.text} ({translation.dest})")
```

```python
# test translate function
translate('house_plant', 'French')
```
